```python
import swiftest
import xarray as xr
from glob import glob
import numpy as np
```

```python
sim = swiftest.Simulation(read_old_output_file=True,simdir="disruption_headon")
```

```python
keep_vars = ['rh','vh','Gmass','radius']
ds1 = sim.data[keep_vars]
ds2 = sim.enc[keep_vars]
ds = xr.combine_nested([ds1,ds2],concat_dim="time")
```
