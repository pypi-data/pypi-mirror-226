```python
import liten
ten = liten.Session()
ten.load_local_demofiles()
spark = ten.spark
agent = ten.agent('syslog')
```

### Linux system log File Example


For looking at linux system log file, we have loaded a sample database. Table syslog contains the data. Extended IETF format include these fields.

Field | Description                            |
------|----------------------------------------|
timestamp | linux os time stamp                  |
hostname  | host that sent the message     |
process   | Name of the thr process. |
pid | process id        |
message | message string for that id |

#### extended ietf format
timestamp hostname process\[pid\]: message header message
#### example log line
Jun 14 15:16:01 combo sshd(pam_unix)\[19939\]: authentication failure; logname= uid=0 euid=0 tty=NODEVssh ruser= rhost=218.188.2.4

Use SQL query to see the number of log file. Look at redirection messages if any.

```python
print(f"Show first three line")
cntDf = ten.spark.sql("select * from syslog limit 3")
cntDf.show()
print(f"Total number of log lines")
cntDf = ten.spark.sql("select count(*) from syslog")
cntDf.show()
```

We can now do these queries on the fly

```python
ten.codegen("Can you tell me how many authentication failures occured in messages from syslog table. The result should be grouped by process field.")
```

Ask to interpret data just by looking at the lines.

```python
ten.codegen("give me only data from syslog table couting the number of errors in last 3 days")
```

```python
df=ten.spark.sql("""
SELECT COUNT(*) AS error_count
FROM syslog
;
""")
df.show()
```

```python
agent.summarize()
```

```python
agent.analyze(df, "Can you see if there have been security violations on this data")
```
